# Student Performance Analysis

## Introduction
In this project, we analyze the performance of students based on various factors such as gender, parental level of education, lunch type, and test preparation course. The dataset includes information on students' scores in math, reading, and writing.

The objective is to identify patterns and correlations that can help understand the factors influencing student performance.

## Data Overview
The dataset consists of the following fields:
- **gender:** Gender of the student
- **race/ethnicity:** Race or ethnicity group
- **parental level of education:** Highest level of education completed by parents
- **lunch:** Type of lunch (standard or free/reduced)
- **test preparation course:** Whether the student completed a test preparation course
- **math score:** Score in math
- **reading score:** Score in reading
- **writing score:** Score in writing

### Dataset Summary
- **Total Records:** 1000
- **Missing Values:** None

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel(r"C:\Users\Habeeb Rahman\Desktop\test_results.xlsx")

sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.boxplot(x='gender', y='math score', data=data, ax=axes[0])
axes[0].set_title('Math Score by Gender')

sns.boxplot(x='gender', y='reading score', data=data, ax=axes[1])
axes[1].set_title('Reading Score by Gender')

sns.boxplot(x='gender', y='writing score', data=data, ax=axes[2])
axes[2].set_title('Writing Score by Gender')

plt.tight_layout()
plt.show()
```
